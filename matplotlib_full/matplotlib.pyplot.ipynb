```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

plt.figure(figsize=(20, 10))

t1 = np.arange(0, 5, 0.05)
t2 = np.arange(0, 5, 0.02)

plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t1, f(t1), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')

plt.show()
```

```python
plt.figure(1)                # the first figure
plt.subplot(211)             # the first subplot in the first figure
plt.plot([1, 2, 3])
plt.subplot(212)             # the second subplot in the first figure
plt.plot([4, 5, 6])


plt.figure(2)                # a second figure
plt.plot([4, 5, 6])          # creates a subplot(111) by default

plt.figure(1)                # figure 1 current; subplot(212) still current
plt.subplot(211)             # make subplot(211) in figure1 current
plt.title('Easy as 1, 2, 3') # subplot 211 title
```

## Working with texts

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
# x is the data, density=1 means make the area under the curve 1, i.e. make probability distribution curve, 
# aplha is the opacity of bars, orientation='horizontal' rotate the graph 90 degrees
n, bins, patches = plt.hist(x, 50, density=1, alpha=0.75, color='r')

plt.xlabel('Smarts', fontsize=20, color='b')
plt.ylabel('Probabilty', fontsize=22, color='g')
plt.title('Histogram of IQ - ' + r'$\sigma_i=15$')
plt.text(44, 0.027, r'$\mu=100,\ \sigma=15$', fontsize=18, color='m')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
ax = plt.subplot(111)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5), arrowprops=dict(facecolor='black', shrink=0.05))

plt.ylim(-2, 2)
plt.show()
```

## Logarithmic and other nonlinear axes

```python
from matplotlib.ticker import NullFormatter
np.random.seed(279)

y = np.random.normal(loc=0.5, scale=0.1, size=10000)
y = y[((y > 0) & (y < 1))]
y.sort()
x = np.arange(len(y))

plt.figure(figsize=(20, 20))

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog')
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)

# format the minor tick on y axis as no marking using Nullformater in ticker class
plt.gca().yaxis.set_minor_formatter(NullFormatter())

# clears the current axes
# plt.cla()

# clears the current figure
# plt.clf()
# And this plots new plot 
# plt.plot(x, y)

plt.show()
```

## setp() - setproperty

## Text Properties

```python
import matplotlib.patches as patches
```

```python
# build a rectangle in axes coords
left, width = -0.25, 1
bottom, height = 0, 1
right = left + width
top = bottom + height

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

# axes coordinates are 0,0 is bottom left and 1,1 is upper right
p1 = patches.Rectangle(
    (left, bottom), width, height,
    fill=False, transform=ax.transAxes, clip_on=False, ec='b', ls='-.'
    )

p2 = patches.Ellipse(
    (0.25, 0.5, 0.5, 0.5), width=2, height=1.5,
    fill=False, transform=ax.transAxes, clip_on=False, ec='r', ls='--')

ax.add_patch(p1)
ax.add_patch(p2)

# ax.text(left, bottom, 'left bottom', transform=ax.transAxes)
ax.text(left, top, 'left top', transform=ax.transAxes)
ax.text(right, top, 'right top', transform=ax.transAxes)
ax.text(right, bottom, 'right bottom', transform=ax.transAxes)

# in vertical alignments if im passing top, its moving downwards relative to default, 
# in horizontal alignments if im passing right, its moving leftwards relative to default
ax.text(left, bottom, 'left bottom', horizontalalignment='right', verticalalignment='top', transform=ax.transAxes)
# ax.text(left, top, 'left top', transform=ax.transAxes)
# ax.text(right, top, 'right top', horizontalalignment='left', verticalalignment='top', transform=ax.transAxes)
# ax.text(right, bottom, 'right bottom',  horizontalalignment='right', verticalalignment='bottom', transform=ax.transAxes)

ax.text((left+right)/2-0.08, (top+bottom)/2+0.1, 'centered', transform=ax.transAxes, fontsize=20, color='g', rotation=45)

ax.set_axis_off()
plt.show()
```
